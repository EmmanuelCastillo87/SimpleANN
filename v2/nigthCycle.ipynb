```python
import pandas as pd
import tensorflow as tf

MEAN_TEMP= 67.872055
MEAN_TIME= 0.536641
STD_TEMP= 14.012810
STD_TIME= 0.069499

coolModelAto = tf.keras.models.load_model('models/coolAto.keras')
warmModelAto = tf.keras.models.load_model('models/warmAto.keras')

def cool(t: float)-> float: 
    t_norm= (t - MEAN_TEMP) / STD_TEMP
    m= (0.51 - MEAN_TIME)/ STD_TIME
    return coolModelAto.predict(pd.DataFrame({'min': [m], 'Ti': [t_norm]})).flatten()[0]

def warm(t: float)-> float: 
    t_norm= (t - MEAN_TEMP) / STD_TEMP
    m= (0.51 - MEAN_TIME)/ STD_TIME
    return warmModelAto.predict(pd.DataFrame({'min': [m], 'Ti': [t_norm]})).flatten()[0]

print(cool(82), ' ', warm(30))
```

```python
model= {
'state': 0,
'TEMPERATURE_IN' : 50,
'TEMPERATURE_OUT' : 81,
'cycles': 0,
}

def changeStage()-> None:
    model['state'] = (model['state'] + 1) % 2

def next(temperature: float)-> float:
    if model['state'] == 0:
        t_next = warm(temperature)
        if t_next > model['TEMPERATURE_OUT']:
            changeStage()
    else:
        t_next = cool(temperature)
        if t_next < model['TEMPERATURE_IN']:
            changeStage()
    return t_next

t = 31
x = range(0, 60*4)
y= []
for i in x:
    y.append(t)
    t = next(t)
```
